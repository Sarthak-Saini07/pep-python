```python

def name1(a,l,i):


  if l == 0:
    return
  if(a[i]=="1"):
    print("One")
  elif(a[i]=="2"):
    print("Two")
  elif(a[i]=="3"):
    print("Three")
  elif(a[i]=="4"):
    print("Four")
  elif(a[i]=="5"):
    print("Five")
  elif(a[i]=="6"):
    print("Six")
  elif(a[i]=="7"):
    print("Seven")
  elif(a[i]=="8"):
    print("Eight")
  elif(a[i]=="9"):
    print("Nine")
  else:
    print("Zero")
  name1(a,l-1,i+1)

st = input("Enter num: ")
name1(st,len(st),0)
```

```python
def my_reduce(lst):
  sum =0
  for i in lst:
    sum = sum+i
  return sum
lst = list(map(int,input("Enter num: ").split()))
print(my_reduce(lst))
```

```python
def dec_to_bin(a):
  if(a == 0):
    return
  rem = a%2
  a = a//2
  dec_to_bin(a)
  print(rem, end="")
dec_to_bin(18)
```
