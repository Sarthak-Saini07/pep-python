```python
n = int(input("Enter a num: ")) ##counting digits
count = 0
while n > 0:
  count = count + 1
  n //= 10
print("count: ",count)
```

```python
n = int(input("Enter a num: ")) ##reversing a number
rev = 0
while n > 0:
  rem = n%10
  rev = rev*10 + rem
  n //= 10
print(rev)
```

```python
n = int(input("Enter a num: ")) ##palindrome
rev = 0
temp = n
while n > 0:
  rem = n%10
  rev = rev*10 + rem
  n //= 10
if (rev == temp):
  print("Palindrome")
else:
  print("Not palidrome")
```
