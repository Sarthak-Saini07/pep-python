```python
n = int(input("Enter number: ")) ## even number
# for i in range(0,n+1,2): ## step method
#   print(i)
for i in range(0,n+1): ## if method
  if(i%2 == 0):
    print(i)
```

```python
n = int(input("Enter number: ")) ## odd number
# for i in range (1,n+1,2): ## step method
#   print(i)
for i in range (0,n+1): ## conditional method
  if(i % 2 != 0):
    print(i)
```

```python
n = int(input("Enter number: ")) ## sum till n numbers
sum = 0
for i in range (0,n+1):
  sum = sum + i
print("Sum : ",sum)
```

```python
n = int (input("Enter a number: ")) ## product till n number
prod = 1
for i in range (n,0,-1):
  prod = prod * i
print(prod)
```

```python
n = int(input("Enter a number: ")) ## count digits using while
inst = 0
while n > 0:
  n = n//10
  inst += 1
print(inst)
```

```python
n = int(input("Enter a number: ")) ## sum of digits using while
sum = 0
while n>0:
  rem = n%10
  sum = sum + rem
  n = n // 10
print(sum)
```

```python
n = int(input("enter a number: ")) ## count digits using for
count = 0
for i in range(n):
  n = n // 10
  count += 1
  if(n == 0):
    break
print(count)
```

```python
n = int(input("enter a number: ")) ## sum of digits using for
count = 0
sum = 0
for i in range(n):
  rem = n%10
  sum += rem
  n = n // 10
  count += 1
  if(n == 0):
    break
print(sum)
```
