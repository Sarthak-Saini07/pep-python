```python
# design a mini social network data model supporting friends, mutual friends and friend suggestions.
social={}
def adding():
  global social
  u = input("Enter name: ")
  social[u]=set()
def adding_friends():
  global social
  u1 = input("Enter name: ")
  u2 = input("Enter name: ")
  if u1 in social:
    social[u1].add(u2)
  if u2 in social:
    social[u2].add(u1)
  else:
    print("User not found")
def mutual():
  global social
  u1 = input("Enter name: ")
  u2 = input("Enter name: ")
  if u1 in social and u2 in social:
    print(social[u1].intersection(social[u2]))
def suggestions():
  global social
  u1 = input("Enter name: ")
  if u1 in social:
    print(social[u1])
  else:
    print("User not found")
def show():
  print(social)
def __main__():
  while True:
    print("1. Add user")
    print("2. Add friends")
    print("3. Mutual friends")
    print("4. Friend suggestions")
    print("5. Show Friends")
    print("6. Exit")
    choice = int(input("Enter choice: "))
    if choice == 1:
      adding()
    elif choice == 2:
      adding_friends()
    elif choice == 3:
      mutual()
    elif choice == 4:
      suggestions()
    elif choice == 5:
      show()
    elif choice == 6:
      break
    else:
      print("Invalid choice")
__main__()
```

```python
# Build a plagiarism checker that compares 2 assignments based on unique words and similarity
# percentage.

a1 = input("Enter Assignment 1: ")
a2 = input("Enter Assignment 2: ")

s1 = set(a1.lower().split())
s2 = set(a2.lower().split())

common = s1 & s2
unique = s1 | s2

print("Common words:", common)
print("Unique words:", unique)

similarity = (len(common) / len(unique)) * 100

print("Similarity = ", similarity, "%")
```
