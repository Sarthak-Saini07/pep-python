```python
for i in range (5):
  for j in range (6):
    print("*" ,end = " ")
  print()
```

```python
for i in range(6):
  # for j in range(1,7):
  print("*"*i,end=" ")
  print()
```

```python
for i in range (6,0,-1):
  print("*"*i,end=" ")
  print()
```

```python
for i in range(1,6):
  for j in range(1,i):
    print(j,end=" ")
  print()
```

```python
for i in range(6,1,-1):
  for j in range(1,i):
    print(j,end=" ")
  print()
```

```python
for i in range(1,6):
  for j in range(1,6):
    print(i,end=" ")
  print()
```

```python
for i in range(1,6):
  for j in range(1,i):
    print(chr(64+j),end=" ")
  print()
```

```python
for i in range(1,6):
  for j in range(0,i):
    print(i,end=" ")
  print()
```
