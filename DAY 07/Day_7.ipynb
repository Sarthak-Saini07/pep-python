```python
def req_of_glass(l,b,n): ## requiring area of glass required for making n no. of tables
  area = l*b
  glass_req = n*area
  return glass_req
a = int(input("Enter num of tables: "))
len = int(input("Enter length: "))
bre = int(input("Enter breadth: "))

print("Required area of glass is: ",req_of_glass(len,bre,a))
```

```python
def simple_interest(p,r,t): ## Simple interest calculator
  si = (p*r*t)/100
  return si
p = int(input("Enter principal: "))
r = int(input("Enter rate: "))
t = int(input("Enter time: "))
print("Simple interest is: ",simple_interest(p,r,t))
```

```python
def temp_conertor(deg): ##temp convertor celsius to fahrenheit
  far = (deg * 1.8) + 32
  return far
deg = int(input("Enter temperature in celsius: "))
print("Temperature in fahrenheit is: ",temp_conertor(deg))
```

```python
def gst_calculator(n):
  if (n>100000):
    return n * 0.18
  elif (n>50000 and n<100000):
    return n * 0.1
  else:
    return n * 0.05

print("GST Calculator")
n = int(input("Enter bill amount: "))
print("Gst on bill is: ",gst_calculator(n))
```

```python
def details (**kwargs):
  for key,value in kwargs.items():
    print(f" {key}:{value}")
data = {}
n = int(input("Enter num: "))
for i in range(n):
  key = input("Enter key: ")
  value = input("Enter value: ")
  data[key]= value
details(**data)
```

```python
def bill_amount(p,q,d):
  price = (p * q) - (p * q * d/100)
  # print("Bill amount is: ",price)

  return price
n = int(input("Enter num: "))
sum = 01
for i in range(n):
  p = int(input("Enter price: "))
  q = int(input("Enter quantity: "))
  d = int(input("Enter discount: "))

  sum = sum + bill_amount(p,q,d)
print ("Total bill : ",sum)

```

```python

def calculator(marks,n):
  sum = 0
  sum = sum + marks
  average = sum / n
  print("Average is: ",average)
  print("Total marks is: ",sum)
  if (average > 30):
    return True
  else:
    return False
name = input("Enter name: ")
n = int(input("Enter number of subjects"))
mark = 0
for i in range(n):
  sub = int(input("Enter marks of sub:"))
  mark = mark + sub

if (calculator(mark,n)):
  print("Pass")
else:
  print("Fail")
```

```python
def calculator(basic):
  hra = 0.5 * basic
  da = 0.6 * basic
  return basic + hra + da
basic = int(input("Enter basic salary: "))
print("Gross salary is: ",calculator(basic))
```
