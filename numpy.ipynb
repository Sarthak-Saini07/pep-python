```python
import numpy as np

students = np.array(["Al", "Bo", "Char", "Dave", "Eva"])

marks = np.array([
    [85, 90, 78, 88],
    [76, 81, 69, 80],
    [92, 89, 95, 91],
    [60, 72, 68, 70],
    [88, 84, 90, 87]
])

average = np.mean(marks, axis=1)
highest = np.max(marks, axis=1)
lowest = np.min(marks, axis=1)
std = np.std(marks, axis=1)
total = np.sum(marks, axis=1)
percentage = total / 400 * 100

ranking = np.argsort(average)[::-1]
print("ranking base: ",ranking)
print("Average:", average)
print("Highest:", highest)
print("Lowest:", lowest)
print("Standard Deviation:", std)
print("Percentage:", percentage)

print("\nRanking")
for i in ranking:
    print(students[i], percentage[i])

class_avg = np.mean(average)
print("\nStudents Above Class Average:")
print(students[average > class_avg])

print("\nSubject-wise Average:")
print(np.mean(marks, axis=0))
```
