```python
class BankAccount:
  def __init__(self,age):
    self.__balances=5000
    self.__age = age
  @property
  def balance(self):
    return self.__balances
  def ages(self):
    return self.__age
  @balance.setter
  def balance(self,value):
    self.__balances=value
acc = BankAccount(22)
acc.balance = 9000
print(acc.balance)
print(acc.ages())
```

```python
from abc import ABC, abstractmethod
class Vehicle(ABC):
  @abstractmethod
  def fuel_type(self):
    pass
class PetrolCar(Vehicle):
  def fuel_type(self):
    print("Petrol")
class ElectricCar(Vehicle):
  def fuel_type(self):
    print("Electric")
p1 = PetrolCar()
p1.fuel_type()
e1 = ElectricCar()
e1.fuel_type()
```

```python
class Restaurant:
    def __init__(self, restaurant_name, cuisine_type):
        self.restaurant_name = restaurant_name
        self.cuisine_type = cuisine_type


class OrderFood(Restaurant):
    def __init__(self, restaurant_name, cuisine_type):
        super().__init__(restaurant_name, cuisine_type)
        self.items = []

    def add_items(self, item):
        self.items.append(item)

    def show_items(self):
        print("Items Ordered:", self.items)


class GenerateBill(OrderFood):
    def __init__(self, restaurant_name, cuisine_type):
        super().__init__(restaurant_name, cuisine_type)

    def generate_bill(self):
        prices = {
            "Pizza": 250,
            "Burger": 150,
            "Pasta": 200,
            "Coke": 50
        }

        total = 0

        print("\n------ BILL ------")
        print("Restaurant:", self.restaurant_name)
        print("Cuisine:", self.cuisine_type)

        for item in self.items:
            print(f"{item} : ₹{prices[item]}")
            total += prices[item]

        print("------------------")
        print("Total Bill: ₹", total)


# Driver Code
obj = GenerateBill("Dominos", "Italian")

obj.add_items("Pizza")
obj.add_items("Coke")
obj.add_items("Burger")

obj.show_items()

obj.generate_bill()

```
