```python
num = input("Enter a number: ")
if "-" in num:
  print ("negative")
else:
  print("positive")
  n = int(num)
  fact = 0
  for i in range(1,n+1):
    if(n % i == 0):
      fact = fact + 1
  if(fact > 2):
    print ("Composite")
  else:
    print("Prime")
```
