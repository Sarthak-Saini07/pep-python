```python
def greet(name, message="Hello"):
  print (f"{message} {name}")
for i in range(5):
  n= input("Enter name: ")
  greet(n)
```

```python
def calc(a,b,op):
  match op:
    case "+":
      return a + b
    case "-":
      return a - b
    case "*":
      return a * b
    case "/":
      return a / b
    case "exp":
      return a**b
    case "cube":
      return pow(a,3)
    case "//":
      return a // b
    case "%":
      return a % b
    case _:
      return "Invalid operator"
num1 = int(input("Enter num: "))
num2 = int(input("Enter num2: "))
ope = input("operator: ")
result = calc(num1,num2,ope)
print(result)
```

```python
def maxi (a,b):
  re = max(a,b)
  return re
num1 = int(input("Enter num: "))
num2 = int(input("Enter num2: "))
result = maxi(num1,num2)
print(result)
```

```python
def eo (a):
  if(a%2==0):
    return "Even"
  else:
    return "Odd"
num1 = int(input("Enter num: "))
result = eo(num1)
print(result)
```

```python
def largest(*arg):
  return max(arg)
a = int(input("Enter number of entries: "))
for i in range (a):
  num = int(input("Enter number: "))
  result = largest(num)
print(result)
```
