```python
value = int(input("Enter value: "))
value2 = int(int(input("Enter value: ")))
op = input("Enter operator:")
if(op == "+"):
  print(value + value2)
elif(op == "-"):
  print(value - value2)
elif(op == "/"):
  print(value / value2)
elif (op == "*"):
    print (value * value2)
elif (op == "%"):
    print(value % value2)
elif (op == "**"):
    print(value ** value2)
elif (op == "//"):
    print(value // value2)
elif (op == "=="):
    print(value == value2)
elif (op == "!="):
    print(value != value2)
elif (op == "<"):
    print(value < value2)
elif (op == "<="):
    print(value <= value2)
elif (op == ">"):
    print(value > value2)
elif (op == ">="):
    print(value >= value2)
else:
  print ("try again")
```
