```python
library = []

def add_book():
    global library
    title = input("Enter title of book: ")
    author = input("Enter author of book: ")
    year = input("Enter year of book: ")

    book = {
        "title": title,
        "author": author,
        "year": year
    }

    library.append(book)
    print("Book added successfully")


def search():
    book = input("Enter book to find: ")

    for i in library:
        if i["title"] == book:
            print(i)
            print("Book found")
            return

    print("Book not found")


def issue():
    book = input("Enter book to issue: ")

    for i in library:
        if i["title"] == book:
            library.remove(i)
            print("Book issued successfully")
            return

    print("Book not found")


def return_book():
    title = input("Enter title: ")
    author = input("Enter author: ")
    year = input("Enter year: ")

    library.append({
        "title": title,
        "author": author,
        "year": year
    })

    print("Book returned successfully")


def display():
    if len(library) == 0:
        print("Library is empty")
    else:
        for i in library:
            print(i)


def backup():
    l2 = library.copy()
    print("Backup created")
    return l2


def restore(l2):
    global library
    library = l2.copy()
    print("Backup restored")


def sort_books():
    library.sort(key=lambda x: x["title"])
    print("Books sorted")


def main():
    l2 = []

    while True:
        print("\n===== LIBRARY MANAGEMENT SYSTEM =====")
        print("1. Add book")
        print("2. Search book")
        print("3. Issue book")
        print("4. Return book")
        print("5. Display books")
        print("6. Backup library")
        print("7. Restore backup")
        print("8. Sort books")
        print("9. Exit")

        ch = int(input("Enter your choice: "))

        if ch == 1:
            add_book()

        elif ch == 2:
            search()

        elif ch == 3:
            issue()

        elif ch == 4:
            return_book()

        elif ch == 5:
            display()

        elif ch == 6:
            l2 = backup()

        elif ch == 7:
            restore(l2)

        elif ch == 8:
            sort_books()

        elif ch == 9:
            print("Thank You")
            break

        else:
            print("Invalid choice")


main()
```
